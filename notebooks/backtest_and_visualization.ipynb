```python
import pandas as pd
import matplotlib.pyplot as plt

# Load one of your trades files
trades = pd.read_csv("../results/AAPL_trades.csv", index_col="entry_date", parse_dates=True)

# Compute cumulative return
cum = (1 + trades["return_pct"]).cumprod() - 1

plt.figure()
plt.plot(cum.index, cum.values)
plt.title("Cumulative P/L for AAPL Reversal Signals")
plt.xlabel("Entry Date")
plt.ylabel("Cumulative Return")
plt.show()
```

A histogram (or box/violin) will tell you if you’re making a few big winners or many small gains—and how fat your tail risk is.

```python
plt.figure()
plt.hist(trades["return_pct"], bins=30)
plt.title("Distribution of Trade Returns")
plt.xlabel("Return per Trade")
plt.ylabel("Count")
plt.show()
```

3. Return vs. Entry Date scatter
Helps you see if your strategy worked better in certain regimes.



```python
plt.figure()
plt.scatter(trades.index, trades["return_pct"], alpha=0.6)
plt.axhline(0, color="k", linestyle="--")
plt.title("Trade Return vs. Entry Date")
plt.xlabel("Entry Date")
plt.ylabel("Return per Trade")
plt.show()
```
