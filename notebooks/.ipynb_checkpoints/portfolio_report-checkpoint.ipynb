```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Load the summary
df_summary = pd.read_csv('../results/portfolio_summary.csv', index_col='ticker')

# 2) Quick look
df_summary.head()

df = pd.read_csv("../results/portfolio_summary.csv", index_col="ticker")

# Bar chart of cumulative returns
plt.figure(figsize=(8,12))
df["cum_return"].sort_values().plot.barh()
plt.title("Cumulative Return by Ticker")
plt.xlabel("Cumulative Return")
plt.tight_layout()
plt.show()

# Scatter of win_rate vs. cum_return
plt.figure()
plt.scatter(df["win_rate"], df["cum_return"])
plt.xlabel("Win Rate")
plt.ylabel("Cumulative Return")
plt.title("Win Rate vs. Cumulative Return")
plt.grid(True)
plt.show()
```
